Il y a quelques mois, je me suis lancé dans un stage en Machine Learning, enthousiaste mais un peu naïf. Très vite, une réalité m’a rattrapé : il ne suffit pas d’écrire du code ou d’entraîner des modèles — il faut aussi savoir collaborer, versionner et documenter son travail proprement. C’est là que Git est devenu, non pas un outil en plus, mais un compagnon de route incontournable.

Git, c’est un peu comme une machine à remonter le temps pour développeurs et data scientists : il garde trace de toutes les modifications de vos projets, permet d’expérimenter sans tout casser, et surtout, facilite la collaboration sur des bases saines. Pourtant, trop souvent, Git est perçu comme un outil réservé aux “codeurs pros”, voire comme un jargon intimidant qu’on repousse à plus tard. Erreur ! Dès qu’on manipule des notebooks, des scripts, des datasets ou des modèles, maîtriser Git devient un superpouvoir.

Dans ce chapitre, je vais vous embarquer dans une exploration accessible et pragmatique de Git, adaptée au monde de la Data Science. Que vous soyez totalement novice ou déjà un peu familier, l’idée est simple : vous donner les clés pour comprendre, utiliser et aimer Git. On va parler de versioning, de branches, de commits, mais aussi de collaboration, de workflows et d’automatisation.

Prêt·e à transformer vos projets en archives bien rangées plutôt qu’en chaos de fichiers nommés script_final_V2_bis_OK.py ? Allez, c’est parti.

  II. Git ≠ GitHub ❗
</h2>

<figure style="padding: 1em; text-align: center;">
    <img src="https://cdn-images-1.medium.com/max/1200/0*oPUR29nEaq80Rhzu" width="500" height="300" alt="Logo Python">

<p style="font-family: 'Segoe UI', Arial, sans-serif; font-size: 1.1em; line-height: 1.6em;">
  Beaucoup pensent à tort que Git et GitHub sont la même chose !  
  <br/><br/>
  <strong>Git</strong> est un système de contrôle de version distribué, installé localement, qui suit chaque modification de vos fichiers.  
  <br/>
  <strong>GitHub</strong> est une plateforme web bâtie sur Git, qui apporte interface graphique, gestion des utilisateurs, Pull Requests, CI/CD et plus encore.  
  <br/><br/>
  Alternatives populaires : GitLab, Bitbucket, SourceTree…
</p>

<h2 style="color: #3498DB; font-family: 'Segoe UI', Arial, sans-serif; font-size: 1.8em; border-bottom: 2px solid #3498DB; padding-bottom: 5px; margin-top: 40px;">
  III. 5 raisons d’utiliser Git en DS/ML 🔍
</h2>
<ol style="font-family: 'Segoe UI', Arial, sans-serif; font-size: 1.1em; margin-left: 1.5em; line-height: 1.6em;">
  <li><strong>Contrôle de versions & récupération</strong> : revenez à n’importe quel état précédent de votre code ou de vos données.</li>
  <li><strong>Collaboration en équipe</strong> : fusionnez les travaux de chacun sans conflits grâce aux branches.</li>
  <li><strong>Reproductibilité</strong> : documentez précisément quelles versions de code et de dataset ont généré vos résultats.</li>
  <li><strong>Innovation & expérimentation</strong> : créez des branches pour tester modèles et pipelines sans toucher au projet principal.</li>
  <li><strong>Automatisation & CI/CD</strong> : intégrez vos workflows dans des pipelines pour tester, déployer et réentraîner vos modèles automatiquement.</li>
</ol>

<h2 style="color: #3498DB; font-family: 'Segoe UI', Arial, sans-serif; font-size: 1.8em; border-bottom: 2px solid #3498DB; padding-bottom: 5px; margin-top: 40px;">
  IV. Terminologie & Workflow Git 📚
</h2>

<figure style="padding: 1em; text-align: center;">
    <img src="https://cdn-images-1.medium.com/max/1200/0*WkzpoNObZ8B7IKP-" width="500" height="300" alt="Logo Python">
</figure>


<ul style="font-family: 'Segoe UI', Arial, sans-serif; font-size: 1.1em; margin-left: 1.5em; line-height: 1.6em;">
  <li><strong>Repository</strong> : le « dépôt » qui contient tout votre projet.</li>
  <li><strong>Branch</strong> : une ligne de développement parallèle.</li>
  <li><strong>Commit</strong> : un snapshot enregistré de vos changements.</li>
  <li><strong>Stage</strong> : l’étape où l’on choisit quels fichiers seront inclus dans le commit.</li>
  <li><strong>Merge</strong> : fusion de deux branches.</li>
  <li><strong>Clone</strong> : copie locale d’un dépôt distant.</li>
  <li><strong>Origin</strong> : alias du dépôt distant d’origine.</li>
  <li><strong>Push</strong> : envoi de vos commits vers le dépôt distant.</li>
  <li><strong>Pull</strong> : récupération des commits distants vers votre local.</li>
  <li><strong>Pull Request</strong> : proposition de fusion pour revue avant intégration.</li>
</ul>

<h2 style="color: #3498DB; font-family: 'Segoe UI', Arial, sans-serif; font-size: 1.8em; border-bottom: 2px solid #3498DB; padding-bottom: 5px; margin-top: 40px;">
  V. Top 10 des commandes Git indispensables 💻
</h2>
<ol style="font-family: 'Segoe UI', Arial, sans-serif; font-size: 1.1em; margin-left: 1.5em; line-height: 1.6em;">
  <li><code>git init</code> – Initialiser un nouveau dépôt.</li>
  <li><code>git clone &lt;url&gt;</code> – Copier un dépôt distant.</li>
  <li><code>git add &lt>fichier&gt;</code> – Mettre en stage vos changements.</li>
  <li><code>git status</code> – Voir l’état de vos fichiers.</li>
  <li><code>git commit -m "message"</code> – Enregistrer un snapshot.</li>
  <li><code>git push</code> – Envoyer vos commits.</li>
  <li><code>git pull</code> – Récupérer les commits distants.</li>
  <li><code>git branch</code> – Lister les branches.</li>
  <li><code>git checkout &lt>branche&gt;</code> – Basculer ou restaurer.</li>
  <li><code>git merge &lt>branche&gt;</code> – Fusionner une branche dans la courante.</li>
</ol>

<figure style="padding: 1em; text-align: center;">

<p style="font-family: 'Segoe UI', Arial, sans-serif; font-size: 1.1em; line-height: 1.6em;">
  Saviez‑vous que Git a été créé en 2005 par Linus Torvalds pour gérer le noyau Linux ?  
  Son nom, « Git », vient de l’argot britannique désignant… un « type désagréable » ! Torvalds plaisantait :  
  <em>« I’m an egotistical bastard, and I name all my projects after myself. First Linux, now Git. »</em>  
  Si vous trouvez Git frustrant, vous savez maintenant pourquoi 😉
</p>
